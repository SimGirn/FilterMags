## FilterMags Test Notebook

The purpose of this notebook is to test the code in ``filtermag.py`` using the test data from SDSS.

```python
import astropy.io.fits as fits
from astropy.table import Table
import os
import filtermag
```

```python
# read in test data
current_dir = os.getcwd()
test_file = os.path.join(current_dir, 'TestData/spec-Spigal-1678-53433-0429.fits')

spectrum_table = Table.read(test_file, 1)
wavelength = 10**spectrum_table['loglam'].data
flux = spectrum_table['flux'].data
```

```python
# run FilterMags
g_mag = filtermag.filter_mag(wavelength, flux, 'sdss_g')
print(g_mag)
```
